```python
datos = datos.fillna(0)
datos
```

```python
datos_a_eliminar = datos.query('Nombre == "Alicia" | Nombre == "Carlos"').index
datos.drop(datos_a_eliminar, axis=0, inplace=True)
datos
```

```python
datos.info()
```

```python
seleccion = (datos['Aprobado'] == 'True') | (datos['Aprobado'] == 'Verdadero')
alumnos_aprobados = datos [seleccion]
```

```python
alumnos_aprobados.to_csv('alumnos_aprobados.csv', index=False)
```

```python
datos['Puntos_extras'] = datos['Nota'].apply(lambda x:x*0.4)
datos['Notas_finales'] = datos['Nota'] + datos['Puntos_extras']
datos['Aprobado_final'] = datos['Notas_finales'].apply(lambda x: 'True' if x >= 7 else 'False')
datos
```
