```python
import numpy as np
```

```python
url = 'https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv'

datos = np.loadtxt(url, delimiter= ',', usecols = np.arange(1,6,1), skiprows= 1)
```

```python
diametro_naranjas = datos[:5000,0]
diametro_toronjas = datos[5000:,0]
peso_naranjas = datos[:5000,1]
peso_toronjas = datos[5000:,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametro_naranjas, peso_naranjas)
plt.plot(diametro_toronjas, peso_toronjas)

#plt.legend([''])
```

```python
n = np.size(datos)
xn = diametro_naranjas
yn = peso_naranjas
an = (n*np.sum(xn*yn)-np.sum(xn)*np.sum(yn)) / (n*np.sum(xn**2) - np.sum(xn)**2)
bn = np.mean(yn) - a*np.mean(xn)
```

```python
xt = diametro_naranjas
yt = peso_naranjas
at = (n*np.sum(xt*yt)-np.sum(xt)*np.sum(yt)) / (n*np.sum(xt**2) - np.sum(xt)**2)
bt = np.mean(yt) - a*np.mean(xt)
```
